```python
def get_market_size(row):
    if row['SLCP'] == "No info":
        return "No info"
    else:
        return row['ClosePrice']*row['SLCP']*1000
def get_EV(row):
    if (row['EV - market_size'] == "No info" or row['Market_size'] == "No info"):
        return "No info"
    else:
        return row['EV - market_size'] + row['Market_size']
def get_EV_EBITDA(row):
    if (row['EV'] == "No info" or row['EBITDA'] == "No info"):
        return "No info"
    else:
        return row['EV']/row['EBITDA']
def PE(row):
    if row['EPS_2018'] == "No info":
        return "No info"
    else:
        return row['ClosePrice']*1000/row['EPS_2018']
def PB(row):
    if (row['Book_value'] == "No info" or row['SLCP'] == "No info"):
        return "No info"
    else:
        return row['ClosePrice']*row['SLCP']*1000/row['Book_value']
stock_data['Market_size'] = stock_data.apply(get_market_size, axis=1)
stock_data['EV'] = stock_data.apply(get_EV, axis=1)
stock_data['EV/EBITDA'] = stock_data.apply(get_EV_EBITDA, axis=1)
stock_data['P/E'] = stock_data.apply(PE, axis=1)
stock_data['P/B'] = stock_data.apply(PB, axis=1)
stock_data
```

```python
def PE(PE):
    if PE == "No info":
        return "No info"
    elif PE <= 5:
        return "<= 5"
    elif PE <= 8:
        return "5 => 8"
    elif PE <= 18:
        return "8 => 18"
    elif PE > 18:
        return "> 18"
stock_data['P/E'] = stock_data['P/E'].apply(lambda x: PE(x))
stock_data.groupby(['P/E'])['ShareCode'].count().to_frame('ShareCode').reset_index()
```

```python
def PB(PB):
    if PB == "No info":
        return "No info"
    elif PB <= 0.5:
        return "<= 0.5"
    elif PB <= 0.8:
        return "0.5 => 0.8"
    elif PB <= 1.5:
        return "0.8 => 1.5"
    elif PB > 1.5:
        return "> 1.5"
stock_data['P/B'] = stock_data['P/B'].apply(lambda x: PB(x))
stock_data.groupby(['P/B'])['ShareCode'].count().to_frame('ShareCode').reset_index()
```

```python
def book_value(book_value):
    if book_value == "No info":
        return "No info"
    elif book_value <= 70000000000:
        return "<= 70 ty"
    elif book_value <= 190000000000:
        return "70 ty => 190 ty"
    elif book_value <= 550000000000:
        return "190 ty => 550 ty"
    elif book_value > 550000000000:
        return "> 550 ty"
stock_data['Book_value'] = stock_data['Book_value'].apply(lambda x: book_value(x))
stock_data.groupby(['Book_value'])['ShareCode'].count().to_frame('ShareCode').reset_index()
```

```python
def market_size(market_size):
    if market_size == "No info":
        return "No info"
    elif market_size <= 50000000000:
        return "<= 50 ty"
    elif market_size <= 150000000000:
        return "50 ty => 150 ty"
    elif market_size <= 600000000000:
        return "150 ty => 600 ty"
    elif market_size > 600000000000:
        return "> 600 ty"
stock_data['Market_size'] = stock_data['Market_size'].apply(lambda x: market_size(x))
stock_data.groupby(['Market_size'])['ShareCode'].count().to_frame('ShareCode').reset_index()
```

```python
def TTS_VCSH(TTS_VCSH):
    if TTS_VCSH == "No info":
        return "No info"
    elif TTS_VCSH <= 1.5:
        return "<= 1.5"
    elif TTS_VCSH <= 2:
        return "1.5 => 2"
    elif TTS_VCSH <= 3:
        return "2 => 3"
    elif TTS_VCSH > 3:
        return "> 3"
stock_data['TTS/VCSH'] = stock_data['TTS/VCSH'].apply(lambda x: TTS_VCSH(x))
stock_data.groupby(['TTS/VCSH'])['ShareCode'].count().to_frame('ShareCode').reset_index()
```
